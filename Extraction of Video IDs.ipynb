```python
# Import required libraries
import time
from selenium import webdriver
import random
from csv import writer
import pandas as pd
from pyvirtualdisplay import Display
import pickle
from webdriver_manager.chrome import ChromeDriverManager
```

```python
def scroll_to_end(driver):
    """
    Function to scroll to the end of the search results page.
    
    Arguments:
    driver -- Selenium WebDriver instance
    
    Returns:
    driver -- Updated Selenium WebDriver instance after scrolling to the end of the page
    """
    # Get the initial height of the page
    prev_ht = driver.execute_script("return document.documentElement.scrollHeight")
    
    # Initialize variables
    i = 0
    sleep_time = 2
    start_time = time.time()
    
    # Loop until the end of the page is reached or a timeout occurs
    while True:
        i += 1
        
        # Print the current iteration number for debugging purposes
        print(i)
        
        # Return the driver if the maximum number of scrolls is reached
        if i == 50:
            return driver
        
        # Print the number of scrolls executed every 100 iterations
        if i % 100 == 0:
            print(i, "scrolls executed")
            
            # Check if the elapsed time exceeds 5.5 hours and break out of the loop if it does
            current_time = time.time()
            if current_time - start_time > (3600 * 5.5):
                print("5 hours elapsed; breaking at i = ", i)
                break
        
        # Scroll to the bottom of the page
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        
        # Wait for a random amount of time between 2 and 3 seconds
        time.sleep(sleep_time + random.uniform(0, 1))
        
        # Get the current height of the page
        ht = driver.execute_script("return document.documentElement.scrollHeight")
        
        # If the page height has not changed after two consecutive checks, increase the wait time by 2 seconds
        if prev_ht == ht:
            time.sleep(2)
            print("Entered same height condition at i = ", i)
            sleep_time += 2
            
            # If the wait time exceeds 20 seconds, break out of the loop
            if sleep_time >= 20:
                print("Sleep time hit 20s; breaking")
                break
        
        # Update the previous height of the page
        prev_ht = ht
        
    # Return the updated driver after scrolling to the end of the page
    return driver

def write_to_txt(driver):
    """
    Function to extract video URLs from the search results page and write them to a text file.
    
    Arguments:
    driver -- Selenium WebDriver instance
    
    Returns:
    None
    """
    # Execute JavaScript code to extract video URLs from the page
    out_list = driver.execute_script("var result = []; " +
        "var all = document.getElementsByTagName('a'); " +
        "for (var i=0, max=all.length; i < max; i++) { " +
        "    if(all[i].getAttribute('id') ==  'video-title')" +                    
        "        result.push(all[i].getAttribute('href')); " +
        "} " +
        " return result; ")
    
    # Write the extracted video IDs to a text file
    with open('Output.txt', 'w') as txtfile: # change the file name to whatever you want
        for item in out_list:
            txtfile.write("%s\n" % item.split("v=")[-1])

if __name__ == "__main__":
    # This code will only be executed if this script is run as the main program (not imported as a module)
    driver = webdriver.Chrome(ChromeDriverManager().install())  # Set up the Selenium WebDriver for Chrome using the ChromeDriverManager package
    driver.get("https://www.youtube.com/results?search_query=novak+djokovic+deportation")  # Navigate to the YouTube search results page for the search query "novak djokovic deportation"
    driver = scroll_to_end(driver)  # Scroll to the end of the page to load all search results
    write_to_csv(driver)  # Write the search result data to a CSV file called "results.csv"
```

```python
# Open the file "Videos.txt" in read mode and assign it to the variable "f"
with open("Videos.txt", "r") as f:
    # Read all the lines of the file and store them in a list called "lines"
    lines = f.readlines()

# Open the file "VideoIDs.txt" in write mode and assign it to the variable "f"
with open("VideoIDs.txt", "w") as f:
    # Loop over each line in the "lines" list
    for line in lines:
        # Find the first occurrence of the "/shorts/" substring in the line
        index = line.find("/shorts/")
        # Continue searching for the "/shorts/" substring until it is no longer found in the line
        while index != -1:
            # Find the next occurrence of the "/shorts/" substring in the line, starting from the previous index
            index = line.find("/shorts/", index + 1)
            # If the "/shorts/" substring is found, extract the video ID from the line and write it to the "VideoIDs.txt" file
            if index != -1:
                value = line[index:].split("/")[2]
                f.write(value + "\n")
```
